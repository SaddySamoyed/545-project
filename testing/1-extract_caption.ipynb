```python
from transformers import AutoTokenizer, pipeline
```

```python
print("Loading the text-generation model, please wait...")
generator = pipeline("text-generation", model="EleutherAI/gpt-j-6B")
print("Model loaded successfully!")
```

```python
def extract_keywords(caption):
    # Define a prompt that instructs the model to extract three keywords.
    prompt = (
        f"Extract three seperate nouns that specify important objects in following caption, separated by commas:\n"
        f"Caption: '{caption}'\nKeywords:"
    )

    # Generate output text using the text-generation pipeline.
    result = generator(prompt, max_length=50, do_sample=False)[0]['generated_text']

    # Post-process the output to get the keywords.
    try:
        # Split on "Keywords:" and take the part after it.
        keywords_text = result.split("Keywords:")[1]
        # Split the text by commas and strip extra spaces.
        keywords = [kw.strip() for kw in keywords_text.split(",") if kw.strip()]
        # Return exactly three keywords (if more are produced, take only the first three).
        return keywords[:3]
    except IndexError:
        return ["No keywords found"]

# Test the function with a sample caption.
caption_example = "A policeman stops on a street with a search dog."
print("Extracted Keywords:", extract_keywords(caption_example))
```
